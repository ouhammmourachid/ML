#### 1 - import the libraries :

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

#### 2 - load the dataset :

```python
data = pd.read_csv("univariate_linear_regression_dataset.csv",names=["X","Y"])
data
```

#### 3 - visualisation dataset:

```python
plt.figure(figsize=(8,6))
plt.title("the visulisation of data of regresion:")
sns.scatterplot(x=data.X,y=data.Y)
plt.xlabel(" the X axis")
plt.ylabel(" the Y axis")
```

```python
SL = stats.linregress(data.X,data.Y)
slope:float = SL.slope
coef_correlation:float= SL.rvalue
intercept :float=SL.intercept
def predict(x:float)->float:
    return slope*x+intercept
```

```python
axes = plt.axes()
axes.grid()
plt.scatter(data.X,data.Y)
fitline = predict(data.X)
plt.plot(data.X,fitline,c='r')
plt.show()
```

```python
x :float = 22.5
f'the prediction of x = {x} is {predict(x)}'
```
