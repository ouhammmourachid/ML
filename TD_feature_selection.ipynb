# Feature Selection:
### 1- using `VarianceThreshold`:

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold
```

```python
df = pd.read_csv('./images_chiffres_codees_niveau_de_gris.csv')
df.head()
```

```python
filter = VarianceThreshold(0.3)
result = filter.fit_transform(df.iloc[:,1:])
```

```python
result
```

```python
X = df.iloc[:,1:]
X.var(axis=0).sort_values(ascending=False)
```

```python
filter.get_support()
```

### 2-using `SelectKBest`:

```python
from sklearn.feature_selection import chi2, SelectKBest
```

```python
X = df.iloc[:,1:]
Y = df.label
chi2(X,Y)
```

```python
filter_2 = SelectKBest(chi2, k=600)
print(X.shape)
fited_data = filter_2.fit_transform(X,Y)
```

```python
fited_data.shape
```
